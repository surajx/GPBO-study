# Guassian Process

#### Random Variable
* A random variable is a function that maps from the sample space to the reals.
$$X: \Omega \rightarrow \mathbb{R}$$

#### Random Process (stochastic process)
* A random process is a collection of random variables index by a set $T$ called the *index set*.
$$\{X_t \mid t\in T\}$$
* Suppose the index set is defined as $T = (-\infty, \infty)$. Now a stochastic process can be thought of as a mapping from the samples space to a space of continuous functions. To see this observe that each r.v. in the stochastic process is now in a one-to-one correspondence with $\mathbb{R}$. This means that a single outcome $\omega \in \Omega$ would trace out a function. Therefore, a single outcome of a random process can be thought of as a function, often called the *sample function*.
$$f(t)\equiv X_t(\omega): T \rightarrow \mathbb{R}$$

#### Gaussian Process (GP)
* A GP is stochastic process with the constraint that any finite sub-collection of r.v. from the stochastic process should be jointly gaussian (a multi-variate gaussian). Formally,
$$\{X_t \mid t\in T\} \text{ is as GP} \iff \forall n\in \mathbb{N}\; \forall t_1\ldots t_n \in T \;\; (X_{t_1},\ldots,X_{t_n}) \sim \mathcal{N}(\;\cdot \mid \mathbf{\mu}, \mathbf{K})$$ 
Where, $\mathbf{\mu} = (\mu(t_1),\ldots, \mu(t_n))$ and $\mathbf{K}_{ij} = k(t_i, t_j)$. Here $\mu(x)$ is the mean function and $k(t, t')$ is the covariance function.


* Since GP is a stochastic process we know that the outcome is a function, say $f(t)$. Then by the definition of GP, $f(t)$ evaluated at some set of points $t_1\ldots t_n$ for some $n\in\mathbb{N}$ should be jointly Gaussian. Let
$$p(\mathbf{f} \mid \mathbf{t}) = \mathcal{N}(\;\mathbf{f}\mid \mathbf{\mu}, \mathbf{K})$$
Where, $\mathbf{f} = (f(t_1),\ldots f(t_n))$ and $\mathbf{t} = (t_1\ldots t_n)$
* Therefore, we can see that a GP defines a prior over functions. The prior belief is captured through the mean  and covariance function.

```python
""" Sample from a gaussian process and plot it """

""" Create Positive Semidefinite Kernel (Covariance Function) """

def k(X_1: np.ndarray, X_2: np.ndarray, kernel='gaussian', args=()):
    def gaussian_kernel(sigma_f=1, l=0.32):
        """ 
            Gaussian Kernel (Squared Exponential)
            @param    sigma_f^2: controls the vertical variation
            @param    l: controls the horizontal length
        """
        # cdist from scipy calculates the pairwise distance between 
        # two collections of objects in n-dimensions
        return sigma_f**2 * np.exp(-(0.5/(l**2))*cdist(X_1, X_2, metric='sqeuclidean'))
    
    def periodic_kernel(amp=1, freq=0.3):
        """ Periodic Kernel """
        return np.exp(-amp*np.sin(2*np.pi*freq*cdist(X_1, X_2, metric='cityblock'))**2)
    
    def bownian_kernel(l=1):
        """Ornstein-Uhlenbeck Kernel depicts brownian motion"""
        return np.exp(-cdist(X_1, X_2, metric='euclidean')/l)
    kernels = {
        'gaussian': gaussian_kernel,
        'periodic': periodic_kernel,
        'brownian': bownian_kernel
    }
    
    
    return kernels[kernel](*args)

def sample_GP(mu, K):
    """ Sample a GP with mean mu and covariance K"""
    N = K.shape[0]
    
    # Colesky only works for Positive Definite matrices, but
    # K is only PSD herefore we need to nudge it a little bit over 0.
    L = np.linalg.cholesky(K + 1e-10*np.eye(N)) # Square root of K
    U = np.random.normal(size=(N,1)) # u ~ N(0, I)
    
    # f ~ \mu + \sigma*(u ~ N(0, I))
    # \sigma is square root of covariance, in the multi-dim case
    # use cholesky decomposition to get the squareroot of a PD matrix
    # f ~ \mu + LU    
    f = mu + np.dot(L, U)
    return f

N=500 # Total number of variables considered from the GP

# Returns evenly spaces numbers over an interval
X = np.linspace(-5, 5, N).reshape(-1,1) # shape = (N, 1)
K = k(X, X, kernel='brownian') # Covariance matrix using gaussian kernel

f = sample_GP(0, K)

pl.plot(X, f, 'b-')
pl.title('A random sample from a GP with Gaussian Kernel')
pl.show()
```

# Linear Regression using GP

### Introduction
We have some data generating process $f(x)$. Due to mesurement error, or due to inherent stochasticity, noise corrupted the output of $f(x)$.


```python
""" Code to Generate Training data """

data_gen_process = lambda x: np.sin(0.9*x).flatten()

def gen_data(data_range=(-5, 5), N=100, noise_stdev=0.05, noise_mean=0):
    """ Generated Noisy data from data generating process """
    noise = np.random.normal(loc=noise_mean, scale=noise_stdev)
    X = np.random.uniform(*data_range, size=(N,1))
    y = data_gen_process(X) + noise
    return (X, y)
```

```python
""" Noise-free Prediction """

#noise less data
X_train, y_train = gen_data(N=10, noise_stdev=0)
X_test = np.linspace(-5, 5, 4).reshape(-1,1)

# K
K_train = k(X_train, X_train)
L_train = np.linalg.cholskey(K_train + 1e-7*np.eyes(len(X_train)))

# K_*
K_train_test = k(X_train, X_test)
L_train_test = np.linalg.cholskey(K_train_test + 1e-7*np.eyes(len(X_train)))

```

```python
""" This is code for simple GP regression. It assumes a zero mean GP Prior """


# This is the true unknown function we are trying to approximate
f = lambda x: np.sin(0.9*x).flatten()
#f = lambda x: (0.25*(x**2)).flatten()


# Define the kernel
def kernel(a, b):
    """ GP squared exponential kernel """
    kernelParameter = 0.1
    sqdist = np.sum(a**2,1).reshape(-1,1) + np.sum(b**2,1) - 2*np.dot(a, b.T)
    return np.exp(-.5 * (1/kernelParameter) * sqdist)

N = 10         # number of training points.
n = 50         # number of test points.
s = 0.00005    # noise variance.

# Sample some input points and noisy versions of the function evaluated at
# these points. 
X = np.random.uniform(-5, 5, size=(N,1))
y = f(X) + s*np.random.randn(N)

K = kernel(X, X)
L = np.linalg.cholesky(K + s*np.eye(N))

# points we're going to make predictions at.
Xtest = np.linspace(-5, 5, n).reshape(-1,1)

# compute the mean at our test points.
Lk = np.linalg.solve(L, kernel(X, Xtest))
mu = np.dot(Lk.T, np.linalg.solve(L, y))

# compute the variance at our test points.
K_ = kernel(Xtest, Xtest)
s2 = np.diag(K_) - np.sum(Lk**2, axis=0)
s = np.sqrt(s2)


# PLOTS:
pl.figure(1)
pl.clf()
pl.plot(X, y, 'r+', ms=20)
pl.plot(Xtest, f(Xtest), 'b-')
pl.gca().fill_between(Xtest.flat, mu-3*s, mu+3*s, color="#dddddd")
pl.plot(Xtest, mu, 'r--', lw=2)
pl.savefig('predictive.png', bbox_inches='tight')
pl.title('Mean predictions plus 3 st.deviations')
pl.axis([-5, 5, -3, 3])

# draw samples from the prior at our test points.
L = np.linalg.cholesky(K_ + 1e-6*np.eye(n))
f_prior = np.dot(L, np.random.normal(size=(n,10)))
pl.figure(2)
pl.clf()
pl.plot(Xtest, f_prior)
pl.title('Ten samples from the GP prior')
pl.axis([-5, 5, -3, 3])
pl.savefig('prior.png', bbox_inches='tight')

# draw samples from the posterior at our test points.
L = np.linalg.cholesky(K_ + 1e-6*np.eye(n) - np.dot(Lk.T, Lk))
f_post = mu.reshape(-1,1) + np.dot(L, np.random.normal(size=(n,10)))
pl.figure(3)
pl.clf()
pl.plot(Xtest, f_post)
pl.title('Ten samples from the GP posterior')
pl.axis([-5, 5, -3, 3])
pl.savefig('post.png', bbox_inches='tight')

pl.show()
```
